```python
from pandas import DataFrame
import json
import pandas as pd
import glob
import os
```

```python
# df1 =pd.read_csv('test_result/chatgpt/*.csv',lineterminator='\n',skipinitialspace=True,) 
# use glob to get all the csv files 
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join('test_result/chatgpt/neg/', "*.csv"))
```

```python
  
# loop over the list of csv files
vertical_concat = pd.DataFrame()
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
#     print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
#     print('Content:')
#     display(df)
    df.drop(df.columns[[0]], axis=1, inplace=True)
    vertical_concat = pd.concat([vertical_concat, df], axis=0)
#     print()
    
display(vertical_concat)
```

```python
yes = vertical_concat.loc[ (vertical_concat['label'] =="Yes") | (vertical_concat['label'] =="Yes.")] 
yes['Tweet'].to_csv('test_result/chatgpt/' + 'all_pos.csv')
yes
```

```python
neg = vertical_concat.loc[ (vertical_concat['label'] !="Yes") & (vertical_concat['label'] !="Yes.")] 
neg['Tweet'].to_csv('test_result/chatgpt/' + 'all_neg.csv')
neg
```
